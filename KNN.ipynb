# **KNN (Classification)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("drive/MyDrive/Colab_Projects/data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
# Droping Useless Features

df.drop(["id", "Unnamed: 32"], axis = 1, inplace= True)
```

```python
df["diagnosis"].value_counts()
```

```python
df["diagnosis"].replace({"M": 0.0, "B": 1.0}, inplace= True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
top_corr = df.corr()["diagnosis"].sort_values(ascending=False)[:31]
top_corr
```

```python
# sns.heatmap(df[top_corr.index].corr(), annot= True, fmt = ".2f", cmap="coolwarm")

# plt.show()
```

```python
X = df[top_corr.index].drop("diagnosis", axis = 1)
y = df["diagnosis"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline(
    [
        ("scaler", StandardScaler()),
        ("model", KNeighborsClassifier())
    ]
)
```

```python
pipeline.fit(X_train, y_train)
```

```python
print("Training Score:", pipeline.score(X_train, y_train))
print("Testing Score:", pipeline.score(X_test, y_test))
```

```python
# Using GridSearchCV

from sklearn.model_selection import GridSearchCV

params = {'model__n_neighbors': [3, 5, 7, 9, 11]}
gscv = GridSearchCV(pipeline, params, cv = 5)
gscv.fit(X_train, y_train)
```

```python
gscv_pred = gscv.predict(X_test)
```

```python
print("Training Score:", gscv.score(X_train, y_train))
print("Testing Score:", gscv.score(X_test, y_test))
```

```python
print(classification_report(y_test, gscv_pred))
```

# **KNN (Regression)**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv("drive/MyDrive/Colab_Projects/student-mat.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df["G3"].value_counts()
```

```python
X = df.drop("G3", axis = 1)
y = df["G3"]
```

```python
categorical_cols = [col for col in X.columns if df[col].dtype == "O"]
numerical_cols = [col for col in X.columns if df[col].dtype == "int64"]
```

```python
[df[categorical_cols[i]].value_counts() for i in range(len(categorical_cols))]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown='ignore', sparse_output = False), categorical_cols),
        ("num", StandardScaler(), numerical_cols)
    ]
)
```

```python
pipeline = Pipeline(
    [
        ("preprocessor", preprocessor),
        ("model", KNeighborsRegressor())
    ]
)
```

```python
pipeline.fit(X_train, y_train)
```

```python
pipeline.predict(X_test)
```

```python
print("Training Score:", pipeline.score(X_train, y_train))
print("Testing Score:", pipeline.score(X_test, y_test))
```
